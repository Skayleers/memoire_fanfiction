# Récupération des textes des fanfictions dans des fichiers .txt

Utilisation de la colonne "body_clean".

### Limitation de longueur
Récupérer l'entièreté des fanfictions pour chaque tag est trop volumineux, je décide de fixer des limites de longueur sur les fanfictions à garder et les récupérer dans un fichier txt pour chaque tag.

```python
# Limites de longueur spécifiques pour chaque tag
length_limits = {
    'Fluff': 12000,
    'Angst': 12000,
    'Hurt/Comfort': 10000,
    'Enemies to lovers': 45000,
    'Friends to lovers': 25000
}
```

```python
# Liste des tags à traiter
tags = ['Fluff', 'Angst', 'Hurt/Comfort', 'Enemies to lovers', 'Friends to lovers']
```

Dictionnaires pour compter le nombre de fanfictions par tag et pour compter le nombre de caractères total par tag

```python
# Dictionnaire pour compter les fanfictions par tag
fanfiction_counts = {tag: 0 for tag in tags}

# Dictionnaire pour compter le nombre total de caractères par fichier
total_characters = {tag: 0 for tag in tags}
```

Pour chaque tag, je récupère les textes des fanfictions appartenant à ce tag et j'exclue celles qui dépasse la limite de caractères fixée précédemment et enregistre la totalité des textes non exclus dans un fichier txt.
(Aussi je remplace le "/" de "Hurt/Comfort" par "_" pour éviter les problèmes de chemin)

```python
# Parcourir chaque tag et enregistrer les textes dans des fichiers séparés
for tag in tags:
    # Remplacer les caractères problématiques dans les noms de fichiers
    safe_tag = tag.replace('/', '_')

    # Filtrer les textes pour le tag actuel
    texts = df[df['tag'] == tag]['body_clean']

    # Obtenir la limite de longueur pour le tag actuel
    max_length = length_limits[tag]

    # Compter et enregistrer les textes qui respectent la limite de longueur
    with open(f'data/{safe_tag}.txt', 'w', encoding='utf-8') as file:
        for text in texts:
            if len(text) <= max_length:
                file.write(text + '\n')
                fanfiction_counts[tag] += 1
                total_characters[tag] += len(text)

    print(f"Les textes pour le tag '{tag}' ont été enregistrés dans '{safe_tag}.txt'.")
```

###Affichage des informations sur chaque tag

```python
# Afficher le nombre de fanfictions par tag qui respectent la limite de longueur
for tag, count in fanfiction_counts.items():
    print(f"Nombre de fanfictions pour le tag '{tag}' faisant moins de {length_limits[tag]} caractères : {count}")

# Afficher le nombre total de caractères par fichier
for tag, chars in total_characters.items():
    print(f"Nombre total de caractères pour le tag '{tag}' : {chars}")
```
