# Lemmatisation

### Fonction de lemmatisation

```python
spacy_pipeline = spacy.load("fr_core_news_sm")
```

```python
# Fonction de lemmatisation avec spaCy
def split_into_lemmas(text):
    doc = spacy_pipeline(text)
    lemmatised_text = [w.lemma_ for w in doc]
    return ' '.join(lemmatised_text)
```

### Lemmatisation des body avec noms des personnages

Je rajoute une colonne 'lemmatised_body' contenant les textes des fanfictions (avec noms des personnages) lemmatisés en appliquant la fonction 'split_into_lemmas' à la colonne 'body_clean".

```python
df_fanfic['lemmatised_body'] = df_fanfic['body_clean'].apply(split_into_lemmas)
```

```python
df_fanfic.head()
```

### Lemmatisation des body sans noms des personnages

Je fais de même pour les textes où les noms de personnages ont été supprimés pour obtenir les textes lemmatisés dans la colonne 'lemmatised_no_chara'.

```python
df_fanfic['lemmatised_no_chara'] = df_fanfic['body_no_chara'].apply(split_into_lemmas)
```
