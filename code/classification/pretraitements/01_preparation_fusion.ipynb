# Suppression des éventuels doublons

```python
#suppression des doublons Fluff
fluff_tab_no_db = fluff_tab.drop_duplicates(subset='work_id', keep='first')
#suppression des doublons Angst
angst_tab_no_db = angst_tab.drop_duplicates(subset='work_id', keep='first')
#suppression des doublons Hurt/Comfort
hc_tab_no_db = hc_tab.drop_duplicates(subset='work_id', keep='first')
#suppression des doublons Enemies to lovers
etl_tab_no_db = etl_tab.drop_duplicates(subset='work_id', keep='first')
#suppression des doublons Friends to lovers
ftl_tab_no_db = ftl_tab.drop_duplicates(subset='work_id', keep='first')
```

# Nombre de fanfictions par tag une fois les éventuels doublons nettoyés

```python
print(f"Nombre de fanfiction Fluff sans doublons : {len(fluff_tab_no_db)}")
print(f"Nombre de fanfiction Angst sans doublons : {len(angst_tab_no_db)}")
print(f"Nombre de fanfiction Hurt/Comfort sans doublons : {len(hc_tab_no_db)}")
print(f"Nombre de fanfiction Enemies to lovers sans doublons : {len(etl_tab_no_db)}")
print(f"Nombre de fanfiction Friends to lovers sans doublons : {len(ftl_tab_no_db)}")
```

```python
print(f"Total des fanfictions collectées : {len(fluff_tab_no_db) + len(angst_tab_no_db) + len(hc_tab_no_db) + len(etl_tab_no_db) + len(ftl_tab_no_db)}")
```

# Rajout d'une colonne tag
Cette colonne servira pour reconnaître les tags associées à chaque fanfiction (utile par exemple pour déterminer les classes lors de l'entrainement des algorithmes de classification).

```python
#colonne tag avec la valeur "Fluff" pour les fanfictions Fluff
fluff_tab_no_db.insert(1, 'tag', 'Fluff')
#colonne tag avec la valeur "Angst" pour les fanfictions Angst
angst_tab_no_db.insert(1, 'tag', 'Angst')
#colonne tag avec la valeur "Hurt/Comfort" pour les fanfictions Hurt/Comfort
hc_tab_no_db.insert(1, 'tag', 'Hurt/Comfort')
#colonne tag avec la valeur "Enemies to lovers" pour les fanfictions Enemies to lovers
etl_tab_no_db.insert(1, 'tag', 'Enemies to lovers')
#colonne tag avec la valeur "Friends to lovers" pour les fanfictions Friends to lovers
ftl_tab_no_db.insert(1, 'tag', 'Friends to lovers')
```

# Enregistrement des nouveaux tableaux

```python
fluff_tab_no_db.to_csv('fluff_fanfics_complet.csv',encoding='utf-8', index=False)
angst_tab_no_db.to_csv('angst_fanfics_complet.csv',encoding='utf-8', index=False)
hc_tab_no_db.to_csv('hc_fanfics_complet.csv',encoding='utf-8', index=False)
etl_tab_no_db.to_csv('etl_fanfics_complet.csv',encoding='utf-8', index=False)
ftl_tab_no_db.to_csv('ftl_fanfics_complet.csv',encoding='utf-8', index=False)
```

# Fusion des tableaux

```python
# Fusionner les DataFrames
fanfics_complet = pd.concat([fluff_tab_no_db, angst_tab_no_db, hc_tab_no_db, etl_tab_no_db, ftl_tab_no_db], ignore_index=True)
fanfics_complet.head()
```
