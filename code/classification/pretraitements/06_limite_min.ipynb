# Analyse des fanfictions les plus courtes

```python
# Calculer le nombre de fanfictions avec moins de 1000 caractères
court_fanfics = df_fanfics[df_fanfics['body_clean'].apply(len) < 1000]

# Afficher le nombre total de fanfictions qui dépassent la limite
total_court_fanfics = len(court_fanfics)
print(f"Nombre total de fanfictions avec moins de 1000 caractères : {total_court_fanfics}")

# Compter le nombre de fanfictions par tag
tag_counts = court_fanfics['tag'].value_counts()

# Afficher le nombre de fanfictions par tag
print("\nNombre de fanfictions par tag qui sont en dessous de la limite :")
print(tag_counts)
```

```python
# Calculer le nombre de fanfictions avec moins de 500 caractères
court_fanfics = df_fanfics[df_fanfics['body_clean'].apply(len) < 500]

# Afficher le nombre total de fanfictions qui dépassent la limite
total_court_fanfics = len(court_fanfics)
print(f"Nombre total de fanfictions avec moins de 500 caractères : {total_court_fanfics}")

# Compter le nombre de fanfictions par tag
tag_counts = court_fanfics['tag'].value_counts()

# Afficher le nombre de fanfictions par tag
print("\nNombre de fanfictions par tag qui sont en dessous de la limite :")
print(tag_counts)
```

# Suppression des fanfictions trop courtes

Je choisis de supprimer les fanfictions dont le nombre de caractères est inférieur à 500. Je ne perds pas beaucoup de données, car elles sont peu nombreuses (12). De plus on peut voir ci-dessus que la plupart de ces fanfictions ne sont en fait pas vraiment des fanfictions, elles ne sont donc pas pertinentes et s'apparenteraient finalement plus à du bruit pour les algorithmes.

```python
# Filtrer les fanfictions avec moins de 500 caractères
df_limit_min = df_fanfics[df_fanfics["nb_caracteres"] >= 500]
```
